```python
import os
import pandas as pd
```

```python
pd.set_option('max_columns',None)
pd.set_option('max_rows', 200)
```

```python
PATH = "/Users/vinodgattani/Documents/Great Learning/WeekII"
FILE = "uberdrive_clean.csv.zip"
```

```python

df = pd.read_csv(os.path.join(PATH, FILE),
                sep=",",
                low_memory=False,
                 parse_dates=[0,1]
                )
```

```python
df.head()
```

```python
df = df[df['trip_duration_in_sec']>0]
```

```python
df['dayofweek'] = df['dayofweek'].astype(int)
df['hour'] = df['hour'].astype(int)
```

### Describe

```python
df.describe(percentiles=[0.9,0.95,0.99])
```

```python
df.describe(include='all')
```

### groupby

https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html

```python
df.groupby(by=['PURPOSE']).agg({"speed":"mean"})
```

```python
df.groupby(by=['PURPOSE']).agg({"speed":["mean","count"],
                               "trip_duration_in_sec":["median"]})
```

```python
grouped = df.groupby(by=['PURPOSE']).agg({"speed":["mean","count"],
                               "trip_duration_in_sec":["median"]})
grouped.sort_values(by=[('speed','count')], ascending = False)
```

### Pivot table

```python
pd.pivot_table(df,
              index=['PURPOSE'],
              columns=['dayofweek'],
              values=['speed'],
              aggfunc=['mean'])
```

```python
pd.pivot_table(df,
              index=['PURPOSE'],
              columns=['dayofweek'],
              values=['speed'],
              aggfunc=['mean', 'count'])
```
